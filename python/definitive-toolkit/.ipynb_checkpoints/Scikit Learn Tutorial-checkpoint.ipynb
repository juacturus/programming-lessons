# train_test_split()

A função ```train_test_split()``` tem como objetivo a separação dos dados em dois datasets diferentes: treinamento e teste. [Documentacao-Oficial](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

## Exemplo 1

````python
# Importando bibliotecas e lendo arquivos
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

root_path = 'C:/Users/thiagoPanini/data-science-courses/handsOn-machineLearning-witch-skLearn-tensorFlow/'
file_path = 'Part-I-The-Fundamentals-of-Machine-Learning/2-End-To-End-Machine-Learning-Project/datasets/housing/'
file_name = 'housing.csv'

df = pd.read_csv(root_path + file_path + file_name)
df.head()
````

Em um algorítimo de Machine Learning, é extremamente importante que o conjunto de dados seja dividido nessas duas partes pois, dessa forma, é possível treinar o modelo com o *conjunto de treinamento* e, logo em seguida, validar o treinamento com o *conjunto de testes*.

Com isso, é possível calcular a eficiência do modelo através de parâmetros como acurácia.

````python
# Dividindo o conjunto de dados em 80% treino e 20% teste
train_set, test_set = train_test_split(df, test_size=.20, random_state=42)

print(f'O conjunto original df possui {df.shape[0]} linhas e {df.shape[1]} colunas.\n')
print(f'O conjunto de TREINO train_set possui {train_set.shape[0]} linhas e {train_set.shape[1]} colunas.')
print(f'O conjunto de TESTE test_set possui {test_set.shape[0]} linhas e {test_set.shape[1]} colunas.')
````

Normalmente, os argumentos passados para a função ```train_test_split()``` são:
    - df = DataFrame a ser splitado;
    - test_size = porcentagem do conjunto de teste a ser atribuída na separação;
    - random_state = garante a randomização dos dados (seed NumPy).
O retorno é uma tupla contendo:
    - 1º elemento = set de treinamento;
    - 2º elemento = set de teste.

# StratifiedShuffleSplit

Uma abordagem avançada sobre a separação dos dados em conjuntos de treinamento e de teste. Com este método, é possível garantir uma maior representatividade dos subconjuntos gerados com relação ao dataset original. [Documentação-Oficial](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html)

## Exemplo 1

````python
# Importando bibliotecas e lendo arquivo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit

root_path = 'C:/Users/thiagoPanini/data-science-courses/handsOn-machineLearning-witch-skLearn-tensorFlow/'
file_path = 'Part-I-The-Fundamentals-of-Machine-Learning/2-End-To-End-Machine-Learning-Project/datasets/housing/'
file_name = 'housing.csv'

df_housing = pd.read_csv(root_path + file_path + file_name)
df_housing.head()
````

Suponha que, conversando com os solicitantes do modelo, seja definido que o atributo **median_income** é extremamente relevante para a predição.

````python
# Observando novamente a distribuição de median_income
df_housing['median_income'].plot(kind='hist', figsize=(9, 6), grid=True, bins=50)
plt.xlabel('median_income')
plt.show()
````

O histograma acima indica que a maior concentração do atributo *median_income* está entre \$20,000 e \$50,000, porém alguns valores vão além de \$60,000. É muito importante que se tenha quantidades suficientes para cada _stratum_ (camada/classe), ou então a estimativa da importância dessa classe pode ser enviesado (deturpado).

    - Isto significa que não é adequado possuir muitas camadas/grupos/strata e cada um destes deve ser largo o suficiente.
    
Em busca da garantia de que este atributo seja representativo, será criada uma nova categoria ["income_cat"] que nada mais é do que um agrupamento dos valores contidos em median_income, aplicadas algumas limitações e transformações para reduzir a quantidade de camadas e aumentar a representatividade destas

````python
# Criando nova categoria
df_housing['income_cat'] = np.ceil(df_housing['median_income'] / 1.5) # Limitando valores
df_housing['income_cat'].where(df_housing['income_cat'] < 5, 5.0, inplace=True)

# Verificando
df_housing['income_cat'].hist(figsize=(10, 5));
````

Referência ```pd.DataFrame.where()```: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.where.html

Dessa forma, é possível aplicar uma amostragem estratificada na categoria *income_cat* através de uma função específica do sickit-learn: ```StratifiedShuffleSplit```

````python
# Aplicando split

split = StratifiedShuffleSplit(n_splits=1, test_size=.2, random_state=42)
for train_index, test_index in split.split(df_housing, df_housing['income_cat']):
    strat_train_set = df_housing.loc[train_index]
    strat_test_set = df_housing.loc[test_index]

# Verificando
strat_test_set['income_cat'].value_counts() / len(strat_test_set)
````

````python
# Por curiosidade, verificando o conteúdo do código acima
print(f'Conteúdo do objeto "split": \n\n{split}')
````

````python
# O que acontece ao aplicar o método .split() do objeto split
print(f'Resultado do método .split() do objeto split: \n\n{split.split(df_housing, df_housing["income_cat"])}')
````

````python
# Simulando o laço for e iterando sobre o generator gerado (train_index e test_index)
for train_index, test_index in split.split(df_housing, df_housing['income_cat']):
    print(f'Conteúdo de train_index: {train_index}')
    print(f'Conteúdo de test_index: {test_index}')
````

````python
# Criação de strat_train_set e strat_test_set que recebem um slicing de df_housing de acordo com os índices dos arraysb acima
for train_index, test_index in split.split(df_housing, df_housing['income_cat']):
    strat_train_set = df_housing.loc[train_index]
    strat_test_set = df_housing.loc[test_index]
strat_train_set.head()
````

````python
# Curiosidade: tamanho de cada um deles
print(f'Tamanho de strat_train_set: {strat_train_set.shape[0]} linhas e {strat_train_set.shape[1]} colunas.')
print(f'Tamanho de strat_test_set: {strat_test_set.shape[0]} linhas e {strat_test_set.shape[1]} colunas.')
````

````python
# São os mesmos valores dos conjuntos de treino e test (train_set e test_set) criados anteriormente
print(f'Tamanho de train_set: {train_set.shape[0]} linhas e {train_set.shape[1]} colunas.')
print(f'Tamanho de test_set: {test_set.shape[0]} linhas e {test_set.shape[1]} colunas.')
````

A grande jogada é que *strat_train_set* e *strat_test_set* são sets criados através de amostras estratificadas, cuja intenção é garantir uma maior representatividade dos conjuntos splitados, aumentando a eficiência e o realismo do modelo. A verificação final da representatividade é feita através da comparação da porcentagem das camadas (strata) da coluna **income_cat** com o conjunto original. 

````python
# Verificando porcentagens - dataset original
df_housing['income_cat'].value_counts() / len(df_housing)
````

````python
# Verificando porcentagens - strat_test_set
strat_test_set['income_cat'].value_counts() / len(strat_test_set)
````

A figura abaixo irá mostrar uma visão geral sobre a representatividade de cada um dos métodos de separação. Quanto mais próximos os valores do dataset original, mais confiável é a separação.

Dessa forma, é possível concluir que os conjuntos de testes gerados através das técnicas de estratificação são muito mais confiáveis no sentido de representar o conjunto como um todo. Por outro lado, as separações obtidas simplesmente por métodos randômicos podem ser levemente enviesadas, tornando um conjunto de testes, por exemplo, não representativo com relação ao modelo e prejudicando os resultados.

````python
# Após isso, é possível remover a coluna adicional criada
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_cat', axis=1, inplace=True)
````

````python
# Apenas por curiosidade, verificando conjuntos
# Teste
strat_test_set.info()
````

````python
# Treino
strat_train_set.info()
````

# Imputer

A classe ```Imputer``` do pacote scikit-learn possui uma aplicação extremamente interessante para tratar dados nulos de determinado conjunto de dados. Através da criação de uma instância da classe Imputer e, especificando o que se deseja calcular do dataset (através do parâmetro ```strategy=```, é possível trabalhar facilmente nesta lacuna. [Documentação-Módulo-preprocessing]([sklearn.preprocessing](http://scikit-learn.org/stable/modules/preprocessing.html)

## Exemplo 1

````python
# Importando bibliotecas, módulos e lendo arquivo
import pandas as pd
import numpy as np
from sklearn.preprocessing import Imputer

root_path = 'C:/Users/thiagoPanini/data-science-courses/handsOn-machineLearning-witch-skLearn-tensorFlow/'
file_path = 'Part-I-The-Fundamentals-of-Machine-Learning/2-End-To-End-Machine-Learning-Project/datasets/housing/'
file_name = 'housing.csv'

df_housing = pd.read_csv(root_path + file_path + file_name)
df_housing.head()
````

````python
# Criando uma instância da classe Imputer
imputer = Imputer(strategy='median')

# Eliminando dados categóricos do DataFrame, visto que iremos calcular a média (mantendo apenas dados numéricos)
housing_num = df_housing.drop('ocean_proximity', axis=1)

# Verificando novo dataset
housing_num.info()
````

````python
# Utilizando método fit() para calcular imputer
imputer.fit(housing_num)
````

Neste momento, o ```imputer``` simplesmente calculou a mediana (definido em ```strategy='median'```) de cada um dos atributos e armazenou o resultado na variável ```statistics_```. A ideia aqui é que, apesar de apenas o atributo *total_bedrooms* ter apresentado dados nulos, não há garantias de que este problema pode se repetir para qualquer outro atributo.

````python
# Verificando variável statistics_
imputer.statistics_
````

````python
# Comparando com dataset original
housing_num.median().values
````

````python
# Transformando o dataset alvo
X = imputer.transform(housing_num)
````

Neste momento, foi criado um novo conjunto de dados **X** contendo a transformação do dataset anterior *housing_num* após a aplicação do imputer.

````python
# Verificando
X
````

````python
# Tranformando dataset em um objeto do tipo DataFrame
df_transformed = pd.DataFrame(X, columns=housing_num.columns)
df_transformed.head()
````

````python
# Verificando quantidades
print(f'O dataset original housing_num possui {housing_num.shape[0]} linhas e {housing_num.shape[1]} colunas.')
print(f'O dataset transformado df_transformed possui {df_transformed.shape[0]} linhas e {df_transformed.shape[1]} colunas.')
````

````python
# Verificando dados nulos
df_transformed.isnull().sum()
````
